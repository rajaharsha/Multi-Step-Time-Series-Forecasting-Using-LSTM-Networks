# Forecasting Peyton Manning Wiki Page Views.

### Loading "Prophet"

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
```

### Loading Wiki Page Views Dataset

```python
df = pd.read_csv('example_wp_peyton_manning.csv')
df['y'] = np.log(df['y'])
df.head()
```

```python
m = Prophet()
m.fit(df);
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```
