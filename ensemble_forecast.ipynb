```python
res = pd.read_csv('84404_ResultsComparision.csv')
res
```

```python
# Standard imports
import numpy as np
import pandas as pd

from bokeh.plotting import figure, output_notebook, show
from bokeh.models import HoverTool, ColumnDataSource

output_notebook()
```

```python
x = [1,2,3,4,5]
actual = list(res['Actual'])
lstm = list(res['LSTM'])
arima = list(res['ARIMA'])
prophet = list(res['Prophet'])
ensemble = list(res['Ensemble'])
```

```python
if 'f' in globals():
    del f

# create a figure object
f=figure(plot_width=700, plot_height=400, title="84404 Actual - GREEN / ARIMA - RED / LTSM - BROWN / Prophet - BLUE / Ensemble - ORANGE")

# create a line plot
f.line(x,actual, line_color="green")
f.line(x,arima,  line_color="red")
f.line(x,lstm,  line_color="brown")
f.line(x,prophet, line_color="blue")
f.line(x,ensemble, line_color="orange")

# show your figure object
show(f)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

'''
rmse=[]
for i in range(len(actual)):
    r = sqrt(mean_squared_error(actual[i], arima[i]))
    rmse.append(r)
'''
arima_rmse = sqrt(mean_squared_error(actual, arima))
lstm_rmse = sqrt(mean_squared_error(actual, lstm))
prophet_rmse = sqrt(mean_squared_error(actual, prophet))
ensemble_rmse = sqrt(mean_squared_error(actual, ensemble))
```

```python
print ('ARIMA        RMSE:      %f' % arima_rmse)
print ('Prophet      RMSE:      %f' % prophet_rmse)
print ('Ensemble     RMSE:      %f' % ensemble_rmse)
print ('LSTM         RMSE:      %f' % lstm_rmse)
```
